```python
import pandas as pd
import numpy as np 
import sklearn 
```

```python
df=pd.read_csv('train.csv')
df.head()
```

```python
x=df.drop('label',axis=1)
y=df['label']
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.1,random_state=44)
```

```python
from sklearn.linear_model import LogisticRegression as lr
lr_model=lr(max_iter=1000)
lr_model.fit(x_train,y_train)
```

```python
predictions=lr_model.predict(x_test)
```

```python
from sklearn import metrics 
c_report=metrics.classification_report(predictions,y_test)
print(c_report)
print(metrics.accuracy_score(predictions,y_test))
```

```python
from sklearn.naive_bayes import MultinomialNB as MNB
nb_model=MNB()
nb_model.fit(x_train,y_train)
```

```python
p_nb=nb_model.predict(x_test)
```

```python
from sklearn import metrics 
c_report=metrics.classification_report(p_nb,y_test)
print(c_report)
print(metrics.accuracy_score(p_nb,y_test))
```

```python
from sklearn.svm import SVC as SVC
svm_model=SVC()
svm_model.fit(x_train,y_train)
```

```python
p_svm=svm_model.predict(x_test)
```

```python
c_report=metrics.classification_report(p_svm,y_test)
print(c_report)
print(metrics.accuracy_score(p_svm,y_test))
```
